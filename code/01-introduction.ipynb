# 4 化学中有大量的数据要分析吗?

**答案是显而易见的**

- 化学和药学都是古董级学科，早已累计了大量的历史数据
- 化学结构拓扑有其自身规律，不同于其它学科的数字、图片、音频、文本数据

- 实验研究向高通量、高内涵方向发展
    - 高通量：大量样品可以平行分析（一次分析）
    - 高内涵：可以获得多维数据，一次分析可以获得样本的多种属性值
    - 分析仪器：LC-MS/MS（高通量）、NMR（高内涵）、高通量活性筛选平台、生物芯片
- 数据特征
    - 数据量大：多行
    - 数据维度高：多列
    - **简单的统计模型难以提炼特征和规律**
    
### 看一下人工智能在化学科研论文中的应用趋势（2020年初统计）：

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/机器学习文献量.csv', encoding='utf-8')
plt.figure(dpi=150)
plt.plot(df['year'],df['machine_learning'],color = 'red', alpha=0.6)
plt.plot(df['year'],df['deep_learning'], color = 'green', alpha=0.6)
plt.legend(['machine learning', 'deep learning'])
plt.show()
```
